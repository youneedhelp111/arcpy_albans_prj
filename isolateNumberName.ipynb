```python
#==================================
# 从属性文本文件提取 Landcover 映射
# 作者: King (Gemini 协作)
#==================================

import json
import os
import re

# ==================== 参数配置 ====================
# 1. 属性文本文件的路径
attr_txt = r"C:\Users\King\Desktop\大三下学习生活\prj-Alban\data\raw\landcover\gaplf2011lc_v30_AL\GAP_LANDFIRE_National_Terrestrial_Ecosystems_2011_Attributes.txt"

# 2. 输出 JSON 的路径
output_json = r"C:\Users\King\Desktop\大三下学习生活\prj-Alban\data\raw\landcover\lc_mapping.json"

# ==================== 核心逻辑 ====================
def parse_attr_txt(file_path, out_j):
    mapping = {}
    print(f"正在读取属性文件: {os.path.basename(file_path)}")
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            
            # 跳过第一行表头
            for line in lines[1:]:
                # 使用正则表达式匹配非空项（处理多个空格或制表符分隔）
                parts = re.split(r'\t| {2,}', line.strip())
                
                # 根据你提供的结构：
                # Value 是第 1 项 (索引 0)
                # ECOLSYS_LU 是倒数第 2 项 (索引 -2)
                if len(parts) >= 19:
                    val_code = int(parts[0])
                    eco_name = parts[-2].strip()
                    
                    # 如果名称为空（如 Value 0），则标记为 No_Data
                    if not eco_name:
                        eco_name = "No_Data_or_Unknown"
                        
                    mapping[val_code] = eco_name
        
        # 导出 JSON
        with open(out_j, 'w', encoding='utf-8') as jf:
            json.dump(mapping, jf, ensure_ascii=False, indent=4)
            
        print(f"--- 提取成功 ---")
        print(f"共提取 {len(mapping)} 条生态系统对应关系。")
        print(f"文件已保存至: {out_j}")

    except Exception as e:
        print(f"处理失败: {e}")

if __name__ == "__main__":
    parse_attr_txt(attr_txt, output_json)
```
